```python
# Read data from excel file  
import pandas as pd 

titanic = pd.read_excel('titanic.xls')
titanic.head()
```

```python
# What percentage of people survived...? 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline

fig, ax = plt.subplots(1, 2, figsize=(15, 8))    # fig(figure), ax(axes) 

titanic['survived'].value_counts().plot.pie(explode=[0, 0.1],
                                            autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Pie plot - Survived')
ax[0].set_ylabel('')

sns.countplot(data=titanic, ax=ax[1], x='survived', hue='survived')
ax[1].set_title('Count plot - Survived')

plt.show()
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))
sns.countplot(data=titanic, ax=ax[0], x='sex', hue='sex')
sns.countplot(data=titanic, ax=ax[1], x='sex', hue='survived')

ax[0].set_ylabel('') # hide y_label 
plt.show
```
